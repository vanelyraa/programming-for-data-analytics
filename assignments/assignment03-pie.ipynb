```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Downloading data from Google Drive
url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"
df = pd.read_csv(url)

# Checking file content
df.head()
```

```python
# Extracting only domains from peoples email adresses using function str.split at delimiter "@" and keeping only the second part of the splitted data, at index 1 which are the domains
df["Domain"] = df["Email"].str.split("@").str[1]

# Sum of each domain
counting_domains = df["Domain"].value_counts()
```

```python
# Creating the pie chart

# Adding blue color palette to chart
colors = ["deepskyblue", "skyblue","dodgerblue"] 

# Find the largest slice to be exploded (separated) for the chart
explode = [0.1 if x == max(counting_domains) else 0 for x in counting_domains]

# Plotting pie chart
# Autopct: add percentages to the plot
# Shadow: add a shadow to the chart, navy edgecolor and darkness changed from 0.7 to 0.8
# Labels = none: removing labels from chart
counting_domains.plot.pie(colors = colors, explode = explode, autopct='%1.1f%%', shadow={"edgecolor": "navy", "shade": 0.9}, labels=None) 

# Title
plt.title("Domains Pie Chart") 

# Displaying labels in legend and positioning it next to the pie not overlapping it
# loc: part of the legend box to be positioned, top-right corner in this case
# bbox_to_anchor: exact coordinates in the graph to position legend box
plt.legend(labels = counting_domains.index, bbox_to_anchor=(0.2,1), loc="upper right", title="Domains")

plt.ylabel("") # Hiding y-axis "count", unnecessary for this plot
plt.tight_layout()
plt.show() 
```
